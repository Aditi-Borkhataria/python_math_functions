4. Calculate the mean and standard deviation of a NumPy array with 100 random integers.

9. Use NumPy to calculate the eigenvalues and eigenvectors of a 3x3 matrix.

```python
import numpy as np
from numpy.linalg import eig
arr = np.array([[2,5,1],
              [4,7,8],
               [6,9,9]])

v,w = eig(arr)
print('eigenvalues:', v)
print('eigenvectors:', w)
```

### MATPLOTLIB

1. Create a line plot with 50 data points that shows the relationship between two variables,
x and y.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,20,50)
y = np.cos(x)

plt.plot(x,y)
plt.xlabel('x data points')
plt.ylabel('y data points')
plt.title('Line graph showing x vs. y')
plt.show()
```

3. Create a bar chart that shows the frequency of occurrence for different categories

```python
import matplotlib.pyplot as plt

categories = ('Category A', 'Category B', 'Category C')
occurences = (34,73,56)
plt.bar(categories, occurences)
plt.xlabel('Categories')
plt.ylabel('Ocurrences')
plt.title('Bar chart showing frequency of occurence for different categories')
plt.show()
```

```python
barWidth = 0.25
fig = plt.subplots(figsize =(10, 8))
 
Math = [90, 60, 62, 100]
Physics = [87, 89,98,92]
Econ = [99, 87, 97, 90]
 
br1 = np.arange(len(Math))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
 
plt.bar(br1, Math, color ='k', width = barWidth,
        edgecolor ='grey', label ='Math')
plt.bar(br2, Physics, color ='b', width = barWidth,
        edgecolor ='grey', label ='Physics')
plt.bar(br3, Econ, color ='y', width = barWidth,
        edgecolor ='grey', label ='Econ')
 
plt.xlabel('Year', fontweight ='bold', fontsize = 12)
plt.ylabel('Grades', fontweight ='bold', fontsize = 12)
plt.xticks([r + barWidth for r in range(len(Math))],
        ['2018', '2019', '2020', '2021'])
plt.title("Grades in school") 
plt.legend()
plt.show()
```

8. Use Matplotlib to create a time series plot that shows the trend in a variable over time

```python
import pandas as pd
data = pd.read_csv(
    "https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/3_TwoNumOrdered.csv", 
    delim_whitespace=True)

# Let's make sure 'date' is actually a date in pandas
data["date"] = pd.to_datetime(data["date"])

date = data["date"]
value = data["value"]
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(date, value);
```

9.  Create a 3D scatter plot that shows the relationship between three variables, x, y, and z.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(10)
y = np.random.rand(10)
z = x*y
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z,
           linewidths=1, alpha=.7,
           edgecolor='k',
           s = 250,
           c=z)
plt.show()
```
